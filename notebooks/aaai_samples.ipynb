# Combination of HuRIC with FrameNet
This notebook analyses the enrichment of the HuRIC dataset with annotated frames coming from FrameNet. The additional frames are used only in train, while for testing only the samples from HuRIC are used.

```python
import numpy as np
import utils

# put there the path to the results you want to show

H_ONLY='../nlunetwork/results/tuning/optimized/h_only/conf_4/huric/modern_right/'
FN_FT='../nlunetwork/results/tuning/optimized/fn_only_hyper_h/huric/modern_right/'
H_FN_FT='../nlunetwork/results/tuning/optimized/h_and_fn_ft/conf_4/huric/with_framenet_ft/'
HURIC_LOCATION = '../data/huric/modern/source'
gold_missing = utils.get_samples_pos_and_lemmas_and_deps(HURIC_LOCATION)
```

```python
# Basic HuRIC results
samples_h = utils.load_json(H_ONLY, 16)
print('#samples_h', len(samples_h))
samples_h_wrong = [s for s in samples_h if s['intent_pred'] != s['intent_true']]
print('#samples_h_wrong', len(samples_h_wrong))
```

```python
# FrameNet in train, HuRIC in test
samples_fn_ft = utils.load_json(FN_FT, 0)
print('#samples_fn_ft', len(samples_fn_ft))
samples_fn_ft_wrong = [s for s in samples_fn_ft if s['intent_pred'] != s['intent_true']]
print('#samples_fn_ft_wrong', len(samples_fn_ft_wrong))
```

```python
# FrameNet + HuRIC train, HuRIC full test
samples_fn_h_ft = utils.load_json(H_FN_FT, 48)
print('#samples_fn_h_ft', len(samples_fn_h_ft))
samples_fn_h_ft_wrong = [s for s in samples_fn_h_ft if s['intent_pred'] != s['intent_true']]
print('#samples_fn_h_ft_wrong', len(samples_fn_h_ft_wrong))
```

```python
utils.add_discriminators(samples_h, gold_missing)
utils.add_discriminators(samples_fn_h_ft, gold_missing)
utils.add_discriminators(samples_fn_ft, gold_missing)

for p1, p2 in zip(sorted(samples_h, key=lambda x: x['file']), sorted(samples_fn_h_ft, key=lambda x: x['file'])):
    lu, _, attn1 = utils.get_intent_attention_arrays(p1)
    _, _, attn2 = utils.get_intent_attention_arrays(p2)
    lu_idx = np.argmax(lu)
    at1_idx = np.argmax(attn1)
    at2_idx = np.argmax(attn2)
    if lu_idx != at1_idx and lu_idx == at2_idx:
        print('true', p1['intent_true'], 'pred_h', p1['intent_pred'], 'pred_fnh', p2['intent_pred'])
        utils.display_sequences(['words', 'lu', 'h->h', 'fn+h->h'], [p1['words'], lu, attn1, attn2])
```

```python
# compare across the three models
for s1, s2, s3 in zip(samples_h, samples_fn_ft, samples_fn_h_ft):
    lu, lu_disc, attn_1 = utils.get_intent_attention_arrays(s1)
    _, _, attn_2 = utils.get_intent_attention_arrays(s2)
    _, _, attn_3 = utils.get_intent_attention_arrays(s3)
    #additional_s = additional_all[s1['id']]
    #additional_s_positions = [el - s1['start_token_id'] for el in additional_s]
    #additional_one_hot = [0 if i not in additional_s_positions else 1 for i in range(s1['length']) ]
    #print(lu, additional_one_hot)
    
    words = s1['words']
    words.insert(0, '<b>true<b>')
    lu.insert(0, '')
    lu_disc.insert(0, '')
    attn_1.insert(0, s1['intent_true'])
    attn_2.insert(0, s2['intent_true'])
    attn_3.insert(0, s3['intent_true'])
    words.insert(1, '<b>pred<b>')
    lu.insert(1, '')
    lu_disc.insert(0, '')
    attn_1.insert(1, s1['intent_pred'])
    attn_2.insert(1, s2['intent_pred'])
    attn_3.insert(1, s3['intent_pred'])

    utils.display_sequences(['words', 'lu', 'disc', 'h->h', 'fn->h', 'fn+h->h'], [words, lu, lu_disc, attn_1, attn_2, attn_3])
    
```
